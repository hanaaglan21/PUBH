## Quiz Questions – Spreadsheets vs Databases (With Answers)

---

### 1. What is the difference between a spreadsheet and a database?

A spreadsheet is a grid-based tool (rows and columns) designed for organizing, calculating, and analyzing relatively small datasets manually.  
A database is a structured system designed to store, manage, and retrieve large amounts of data efficiently using relationships and queries.

---

### 2. What are the advantages and disadvantages of using a spreadsheet for health data management?

**Advantages:**
- Easy to use and learn
- Good for small datasets
- Useful for quick calculations and visualizations
- Ideal for individual use

**Disadvantages:**
- Prone to human error
- Limited scalability
- Weak security controls
- Poor handling of large, complex, or multi-user data

---

### 3. What are the advantages and disadvantages of using a database for health data management?

**Advantages:**
- Handles large datasets efficiently
- Supports multiple users simultaneously
- Stronger security and access controls
- Reduces redundancy through structured design
- Supports complex queries

**Disadvantages:**
- More complex to design and manage
- Requires technical knowledge (e.g., SQL)
- Higher setup and maintenance costs

---

### 4. What are some examples of tasks that can be performed using a spreadsheet?

- Calculating averages or totals (e.g., patient blood pressure averages)
- Creating charts and graphs
- Filtering or sorting small datasets
- Budget tracking
- Simple data entry and reporting

---

### 5. What are some examples of tasks that can be performed using a database?

- Storing millions of patient records
- Running complex SQL queries
- Managing hospital electronic health records (EHR)
- Enforcing role-based access controls
- Linking multiple tables (patients, visits, medications)

---

### 6. What are some examples of tasks that can be performed using both a spreadsheet and a database?

- Storing patient demographic data
- Sorting and filtering records
- Generating reports
- Basic data analysis
- Importing and exporting datasets

---

### 7. What are some examples of tasks that can be performed using neither a spreadsheet nor a database?

- Performing physical medical procedures
- Conducting face-to-face patient counseling
- Real-time emergency medical treatment
- Conducting laboratory experiments manually

---

### 8. What are some examples of tasks that can be performed using a spreadsheet but not a database?

- Creating quick ad-hoc financial models using built-in formulas
- Performing manual cell-by-cell calculations interactively
- Creating simple visual dashboards without programming

(Note: Databases can export data for visualization, but they do not natively provide spreadsheet-style formula manipulation.)

---

### 9. What are some examples of tasks that can be performed using a database but not a spreadsheet?

- Handling millions of interconnected health records
- Enforcing user-level security permissions
- Performing multi-table relational joins
- Managing concurrent multi-user access
- Maintaining data integrity constraints

---

### 10. What are some examples of tasks that can be performed using both a spreadsheet and a database?

- Storing tabular health data
- Sorting and filtering information
- Generating summary reports
- Exporting data for analysis
- Performing basic statistical calculations
