# Data Visualization – Quiz Questions

## Setup
Import the packages used in the notebook.

```python
import pandas
import seaborn
from matplotlib import pyplot
```

## Load the data
Use the dataframe name `df` like the notebook does.

```python
df = pandas.read_csv('https://raw.githubusercontent.com/juanklopper/TutorialData/main/heart_failure.csv')
df.head()
```

## Question 1
How do you import the Seaborn package in Python?

```python
import seaborn
```

## Question 2
How do you create a scatter plot in Seaborn for variables x and y in a DataFrame df?

```python
seaborn.scatterplot(
    data=df,
    x='age',
    y='platelets'
)
```

## Question 3
How do you create a line plot in Seaborn for variables x and y in a DataFrame df?

```python
seaborn.lineplot(
    data=df,
    x='x',
    y='y'
)
```

## Question 4
How do you create a histogram in Seaborn for a variable x in a DataFrame df?

```python
seaborn.displot(
    df,
    x='age'
)
```

## Question 5
How do you create a stacked histogram in Seaborn for variables x and hue y in a DataFrame df?

```python
seaborn.displot(
    df,
    x='x',
    hue='y',
    multiple='stack'
)
```

## Question 6
How do you create a heatmap in Seaborn for variables x and y in a DataFrame df?

```python
seaborn.displot(
    df,
    x='x',
    y='y',
    cbar=True
)
```

## Question 7
How do you create a box plot in Seaborn for variables x and y in a DataFrame df?

```python
seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    kind='box'
)
```

## Question 8
How do you create a violin plot in Seaborn for variables x and y in a DataFrame df?

```python
seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    kind='violin'
)
```

## Question 9
How do you create a bar plot in Seaborn for variables x and y in a DataFrame df?

```python
seaborn.barplot(
    data=df,
    x='x',
    y='y'
)
```

## Question 13
How do you change the color palette of a Seaborn plot to 'Blues'?

```python
seaborn.set_palette('Blues')
```

```python
ax.set_yscale('log')
```

## Question 15
How do you add a regression line to a scatter plot in Seaborn for variables x and y in a DataFrame df?

```python
seaborn.regplot(
    data=df,
    x='x',
    y='y'
)
```
