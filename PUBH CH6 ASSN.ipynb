## Question 1  
Use Python to calculate the expression:

3 × 8^33 + 7^5

```python
3 * 8**33 + 7**5
```

### Explanation

- `**` is the exponent operator in Python.
- `8**33` calculates 8 raised to the 33rd power.
- `7**5` calculates 7 raised to the 5th power.
- Python handles very large integers automatically.

## Question 4  

Create the variable `i` and assign it the value 1.  
Create a loop that prints `i²`.  
Increment `i` by 1 until `i²` exceeds 1000.  
Print the final value of `i`.

```python
i = 1

while i**2 <= 1000:
    print(i**2)
    i += 1

print("Final value of i:", i)
```

### Explanation

The loop continues while:

i² ≤ 1000

31² = 961 (still less than 1000)  
32² = 1024 (greater than 1000)

Therefore, the loop stops when i = 32.

Final answer: i = 32
