## Task 3
Reverse the order of the set difference in the code cell above and determine if the set difference between the objects `set_A` and `set_B` is the same as the set difference between `set_B` and `set_A`.

```python
set_A = {'a', 'b', 'c', 'd', 'e', 'f'}
set_B = {'a', 'c', 'f', 'g'}

set_A.difference(set_B)
```

```python
set_B.difference(set_A)
```

### Answer
No, the set differences are **not the same**.

- `set_A.difference(set_B)` gives the elements in `set_A` that are not in `set_B`
- `set_B.difference(set_A)` gives the elements in `set_B` that are not in `set_A`

So:

- `set_A.difference(set_B)` → `{'b', 'd', 'e'}`
- `set_B.difference(set_A)` → `{'g'}`

This shows that set difference depends on the order.
