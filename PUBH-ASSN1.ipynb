## Task 1: Another example of the misuse of health data

Another well-documented example of the misuse of health data is the **Tuskegee Syphilis Study** conducted by the U.S. Public Health Service between 1932 and 1972. In this study, Black men in Tuskegee, Alabama were recruited to observe the natural progression of untreated syphilis. Participants were not informed of their diagnosis, nor were they told that effective treatment (penicillin) became available during the study. Instead, treatment was deliberately withheld, even when it was known to be lifesaving.  

The ethical violations were severe, including lack of informed consent, deception, exploitation of a vulnerable population, and intentional harm. The study caused preventable suffering and deaths and contributed to long-standing mistrust of the healthcare system among Black communities. This misuse could have been prevented through genuine informed consent, independent ethical oversight, respect for participant autonomy, and the intervention of an Institutional Review Board, which did not exist at the time.

## Task 2: Another example of a breach in health data security

A major health data breach occurred in **2017 involving Equifax**, a credit reporting agency that stored sensitive health-related insurance and employment data for millions of individuals. Hackers exploited a known vulnerability in the company’s software, gaining access to personal information—including Social Security numbers, dates of birth, and addresses—of approximately **147 million people**.  

While Equifax is not a healthcare provider, the breach affected health insurance records and exposed information commonly used in healthcare administration. The consequences included identity theft, financial fraud, and widespread public outrage, as well as legal settlements exceeding **$700 million**. This breach could have been prevented through timely software updates, stronger cybersecurity practices, encryption of sensitive data, and proactive monitoring for intrusions.

## Task 3: Another example of bias in health data

An additional example of bias in health data can be found in **cardiovascular disease research**, where women have historically been underrepresented in clinical trials. For decades, diagnostic criteria and treatment guidelines for heart disease were largely based on data from male patients, leading to misdiagnosis or delayed diagnosis in women.  
+
Women often present different symptoms—such as fatigue, nausea, or jaw pain rather than chest pain—which were not adequately reflected in medical training or datasets. As a result, women have experienced higher mortality rates following heart attacks. This bias could have been prevented by ensuring diverse and representative data collection and requiring sex-disaggregated analysis in clinical trials.

## Task 4: Example of anonymization of health data

A common example of anonymization of health data occurs in **large-scale public health research datasets**, such as those released by national health agencies for epidemiological studies. Patient-level hospital admission data may be anonymized using **pseudonymization and generalization**, where direct identifiers are removed and replaced with random IDs, while ages are converted into ranges.  

Many datasets also apply **k-anonymity**, ensuring that no individual can be distinguished from at least *k − 1* others based on shared attributes.

## Task 5: The George Washington University Office of Human Research

The **George Washington University Office of Human Research (OHR)** oversees the ethical conduct of research involving human subjects at GWU. The office manages the Institutional Review Board (IRB) process, reviewing research protocols to ensure compliance with federal regulations and ethical principles.  

Its primary role is to protect the rights, safety, and welfare of research participants by evaluating informed consent procedures, assessing risks and benefits, and ensuring privacy and confidentiality protections are in place.




## Quiz Questions – Ethics and Privacy 

---

### **1. What is the primary goal of data anonymization?**
To protect individuals’ privacy by removing or altering identifying information so data cannot be linked back to a specific person, while still allowing useful analysis.

---

### **2. What is the Havasupai Tribe case about?**
It concerns the misuse of genetic data when researchers used blood samples collected for diabetes research to conduct other studies without the tribe’s informed consent, violating ethical and cultural rights.

---

### **3. What was the largest health data breach in history?**
The **Anthem Inc. breach (2015)**, which exposed the personal data of approximately **78.8 million people**.

---

### **4. What is the pulse oximeter controversy about?**
It involves racial bias in pulse oximeters, which were found to be less accurate for people with darker skin tones, leading to missed or delayed detection of low blood oxygen levels.

---

### **5. What is the role of an Institutional Review Board (IRB) in research?**
To review, approve, and monitor research involving human subjects to ensure ethical standards are met and participants’ rights, safety, and privacy are protected.

---

### **6. What is the difference between spreadsheet software and database software for health data management?**
Spreadsheets are best for small, simple datasets and manual analysis, while databases are designed for large, complex datasets with multiple users, better security, and structured querying.

---

### **7. What is a conflict of interest in public health research?**
A situation where a researcher’s personal, financial, or professional interests could influence or bias their research decisions or outcomes.

---

### **8. What is informed consent in the context of health data collection?**
It is when individuals are fully informed about how their data will be collected, used, stored, and shared—and voluntarily agree to participate.

---

### **9. What is data masking in the context of de-identification of health data?**
The process of hiding or removing identifiable information (such as names or IDs) by replacing it with symbols, random characters, or blanks.

---

### **10. What is the principle of k-anonymity in the context of de-identification of health data?**
It ensures that each individual in a dataset cannot be distinguished from at least **k − 1** other individuals based on shared characteristics.

---

### **11. What is the main concern about conflicts of interest in public health research?**
They can compromise research integrity, bias results, and undermine public trust.

---

### **12. What is the first step in managing conflicts of interest in public health research?**
Disclosure—researchers must openly declare any potential conflicts of interest.

---

### **13. What is the role of independent oversight in managing conflicts of interest in public health research?**
To objectively review research decisions and ensure conflicts do not improperly influence study design, analysis, or conclusions.

---

### **14. What is one strategy for mitigating conflicts of interest in public health research?**
Requiring independent data analysis or excluding conflicted individuals from key decision-making roles.

---

### **15. What is the potential impact of not properly managing conflicts of interest in public health research?**
Biased findings, harmful public health decisions, loss of credibility, and reduced trust in scientific institutions.
